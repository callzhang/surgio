```python
import yaml
file = "derek-list-clash.conf"
# file = 'xutong-clash.conf'
with open(file, "r") as stream:
    try:
        server_dict = yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)
```

```python
import pandas as pd

server = pd.DataFrame(server_dict['proxies'])
print(server.columns)
server.sample(5)
```

```python
server.rename({
    'name': 'nodeName',
    'server': 'hostname',
    'udp': 'udp-repay',
    'cipher': 'method',
    'ws-path': 'path',
    'skip-cert-verify': 'skipCertVerify'
}, axis=1, inplace=True)
# server.drop('method', axis=1, inplace=True)
server_dict = server.to_dict(orient='records')
server_dict
```

```python
def print_dict(d, indent=0):
    ind = "\t"*indent
    print('{')
    for k, v in d.items():
        print(f'{ind}"{k}": ', end='')
        if isinstance(v, bool):
            print(f'{"true" if v else "false"},')
        elif isinstance(v, (str, int)):
            print(f'"{v}",')
        elif type(v) is dict:
            print_dict(v, indent+1)
        else:
            raise Exception(f'Unknown type: {type(v)}')
    print("\t"*(indent-1)+'},')
for kv in server_dict:
    print_dict(kv, 0)

```
